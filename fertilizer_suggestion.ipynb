```python
n=int(input("Enter Nitrogen"))
p=int(input("Enter Phosphorus"))
k=int(input("Enter Potassium"))
def suggestion_fertilizer(n,p,k):
    suggestions = []
    if n < 40:
        suggestions.append("Nitrogen is low → Use Urea")
    elif n > 80:
        suggestions.append("Nitrogen is high → Avoid Urea")

    if p < 30:
        suggestions.append("Phosphorus is low → Use DAP")
    elif p > 60:
        suggestions.append("Phosphorus is high → Avoid DAP")

    if k < 30:
        suggestions.append("Potassium is low → Use MOP")
    elif k > 60:
        suggestions.append("Potassium is high → Avoid MOP")

    if len(suggestions) == 0:
        suggestions.append("NPK levels are balanced. No major fertilizer needed.")

    return suggestions
```

```python
print(suggestion_fertilizer(n,p,k))
```
