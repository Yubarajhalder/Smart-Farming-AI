```python
import pandas as pd
import joblib


# Load Models
crop_model = joblib.load("crop_model.pkl")
soil_model = joblib.load("soil_health_model.pkl")


# Fertilizer Suggestion Function
def fertilizer_suggestion(n, p, k):

    suggestions = []

    if n < 40:
        suggestions.append("Nitrogen is low → Use Urea")

    elif n > 80:
        suggestions.append("Nitrogen is high → Avoid Urea")


    if p < 30:
        suggestions.append("Phosphorus is low → Use DAP")

    elif p > 60:
        suggestions.append("Phosphorus is high → Avoid DAP")


    if k < 30:
        suggestions.append("Potassium is low → Use MOP")

    elif k > 60:
        suggestions.append("Potassium is high → Avoid MOP")


    if len(suggestions) == 0:
        suggestions.append("NPK levels are balanced.")

    return suggestions


# User Input
N = int(input("Enter Nitrogen: "))
P = int(input("Enter Phosphorus: "))
K = int(input("Enter Potassium: "))
temperature = float(input("Enter Temperature: "))
humidity = float(input("Enter Humidity: "))
ph = float(input("Enter pH: "))
rainfall = float(input("Enter Rainfall: "))


# Crop Prediction Input
crop_data = pd.DataFrame([{
    "N": N,
    "P": P,
    "K": K,
    "temperature": temperature,
    "humidity": humidity,
    "ph": ph,
    "rainfall": rainfall
}])


# Soil Health Input
soil_data = pd.DataFrame([{
    "N": N,
    "P": P,
    "K": K,
    "ph": ph
}])


# Predictions
crop_prediction = crop_model.predict(crop_data)[0]

soil_prediction = soil_model.predict(soil_data)[0]

fertilizer_prediction = fertilizer_suggestion(N, P, K)


# Final Output
print("\n===== SMART FARMING RESULT =====")

print("\nRecommended Crop:")
print(crop_prediction)

print("\nSoil Health:")
print(soil_prediction)

print("\nFertilizer Suggestions:")
for item in fertilizer_prediction:
    print("-", item)
```
